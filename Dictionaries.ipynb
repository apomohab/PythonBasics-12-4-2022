```python
# Write dictionary

dict1 = {"Name":"Saeed", "Age":"29" , "Job":"Cooder"}
dict2 = {"Salary":150 , "Prices":300}
print(dict1)
print(dict2)
```

```python
#Called Keys
print(dict1["Name"])
print(dict2["Salary"])
print(dict2["Prices"])
```

```python
#Assign new value

dict1["Country"] = "Egypt"
dict1["weight"] = "80"
print(dict1["Country"])
print(dict1["weight"])
```

```python
#Summ values

dict2["Salary"] = dict2["Salary"] + 40
print(dict2["Salary"])

dict2["PRices"] = dict2["Prices"] + 100
dict2
```

```python
#Add values inside lists

Ages = {"Sara":[10,15,20], "Mohab":[20,30,40]}
Ages
print(Ages["Sara"])

Ages['Sara'] = Ages["Sara"] + Ages["Mohab"]
print(Ages)

print(Ages["Sara"][2])
```

```python
#Nested dictionary - same lists bud curly brakets

employees = {"Ahmed":{"Salary":50 , "Age":25} , "Mohab":{"Salary":150 , "Age":20}}
print(employees["Ahmed"]["Salary"])
print(employees["Mohab"]["Age"])
```

```python
#Print keys only and values only - all

print(dict1.keys())
print(dict2.keys())
print(dict2.values())
print(dict2.values())

```
